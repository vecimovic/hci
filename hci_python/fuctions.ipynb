## Funkcija za crtanje polarnih funkcija

```python
def draw_fun(fun, period=2*np.pi):
    plt.axes(projection = 'polar')
    thetas = np.arange(0, period, 0.005)
    f = sp.lambdify(x, fun, "numpy")
    r = []
    for i in thetas:
        r.append(f(i))
    plt.polar(thetas, r, 'g.')
    plt.show()
    
```

## Easy funkcije

```python
def easy_fun1():
    return 2 + 0 * x

draw_fun(easy_fun1())
get_function_length(easy_fun1())
```

```python
def easy_fun2():
    return 0.2 - 2 * sp.sin(x)

draw_fun(easy_fun2())
get_function_length(easy_fun2())
```

```python
def easy_fun3():
    return 2 + 0.3 * sp.sin(2 * x)

draw_fun(easy_fun3())
get_function_length(easy_fun3())
```

```python
def easy_fun4():
    return 0.7 * x

draw_fun(easy_fun4())
get_function_length(easy_fun4())
```

## Medium funkcije

```python
def medium_fun1():
    return 2 - 2 * sp.sin(2 * x) + sp.cos(4 *x)

draw_fun(medium_fun1())
get_function_length(medium_fun1())
```

```python
def medium_fun2():
    return 2 + 2 * sp.sin(3 * x)

draw_fun(medium_fun2())
get_function_length(medium_fun2())
```

```python
def medium_fun3():
    return 3 - 0.8 * sp.sin(3 * x) * sp.cos(6 * x)

draw_fun(medium_fun3())
get_function_length(medium_fun3())
```

```python
def medium_fun4():
    return 3 - 1.5 * sp.sin(4 * x) * sp.cos(1 * x)

draw_fun(medium_fun4())
get_function_length(medium_fun4())
```

## Hard funckije

```python
def hard_fun1():
    return 1 + 2* sp.sin(7 * x) * sp.cos(3 * x)

draw_fun(hard_fun1())
get_function_length(hard_fun1())
```

```python
def hard_fun2():
    return 1.8 + 2 * sp.sin(6 * x)

draw_fun(hard_fun2())
print("Lenght: ", get_function_length(hard_fun2()))
```
